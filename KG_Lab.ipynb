```python
import pandas as pd

df = pd.read_csv('filtered_data.csv')
corpus_sentences = df.corpus_sentences.to_list()
filtered_prefix = df.filtered_prefix.to_list()
filtered_original_sentences = df.filtered_original_sentences.to_list()
filtered_uu_text = df.filtered_uu_text.to_list()
df
```

```python
from bertopic import BERTopic
from sentence_transformers import SentenceTransformer, util
import torch

all_embeddings =  torch.load("/content/multilingual-e5-large-instruct.pt", map_location="cpu").numpy()

embedding_model = SentenceTransformer('intfloat/multilingual-e5-large-instruct',
                                      prompts={"query": "Instruct: Identify the area of law or legal topic of paragraphs based on their content\nQuery: "},
                                     default_prompt_name="query")

topic_model = BERTopic.load("/content/")
topic_model.get_topic_info()
```

```python
import re
from tqdm import tqdm
import nltk
from nltk.corpus import stopwords

topic_info = topic_model.get_topic_info()

nltk.download('stopwords')
stop_words = set(stopwords.words('indonesian'))

def inferece_new_document(document):
  # Filtering
  text = " ".join(re.sub(r"[^a-zA-Z\s]", " ", document).split())
  pattern = r'\b(nol|satu|dua|tiga|empat|lima|enam|tujuh|delapan|sembilan|sepuluh|sebelas|belas|puluh|seratus|ratus|ribu|juta|miliar|triliun)\b'
  text = " ".join(re.sub(pattern, " ", text, flags=re.IGNORECASE).split())
  text = ' '.join([word for word in text.split() if word not in stop_words])

  if text == "":
    return topic_info['Name'].to_list()[0], 100
  else:
    embeddings = embedding_model.encode(text, normalize_embeddings=True).reshape(1, 1024)
    topics, probs = topic_model.transform([text],embeddings)
    return topic_info['Name'].to_list()[topics[0]+1], probs[0]
```

```python
def get_document_by_topic(input_topics):
  selected_topic = input_topics.split("_")[0]
  documents = document_info.loc[document_info["Topic"]==int(selected_topic) ]
  return {output_documents: documents }
```

```python

visualize_topics = topic_model.visualize_documents(corpus_sentences, embeddings=all_embeddings, hide_annotations=True)

# hierarchical_topics = topic_model.hierarchical_topics(corpus_sentences)
# fig_hierarchy = topic_model.visualize_hierarchy(hierarchical_topics=hierarchical_topics)

topic_info = topic_model.get_topic_info()
document_info = topic_model.get_document_info(filtered_uu_text).drop(columns=['Representation', 'Representative_Docs','Top_n_words','Representative_document'])
document_info['Name'] = document_info['Name'].apply(lambda x: x.replace('_', ' ', 1).replace('_', ''))

# topic_info.to_csv('topic_info.csv', index=False)
document_info.to_csv('document_info.csv', index=False)

def overview_visualization():

  return {
      output_visualize_topics: visualize_topics,
      # output_visualize_hierarchy: fig_hierarchy,

      }

def download_document_info():
  return 'document_info.csv'
```
